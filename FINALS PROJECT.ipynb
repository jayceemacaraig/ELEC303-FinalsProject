<h2>IMPORTING IMPORTANT LIBRARIES</h2>

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('datasets\\Family Income and Expenditure.csv')
```

```python
data
```

```python
data.columns
```

<hr><br>
<h2>CLEANING DATA TYPES AND FILLING UP NULL VALUES</h2>

```python
df = data[['Region','Total Number of Family members','Agricultural Household indicator','Members with age less than 5 year old','Total Household Income', 'Total Food Expenditure', 'Bread and Cereals Expenditure', 'Total Rice Expenditure', 'Meat Expenditure', 'Total Fish and  marine products Expenditure','Fruit Expenditure','Vegetables Expenditure', 'Alcoholic Beverages Expenditure']]

df
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
# Derived metrics
df['Rice_Share'] = np.round((df['Total Rice Expenditure'] / df['Total Food Expenditure']) * 100,0)
df['Protein_Share'] = np.round(((df['Meat Expenditure'] + df['Total Fish and  marine products Expenditure']) / df['Total Food Expenditure'])*100,0)
df['Veggie_Fruit_Share'] = np.round(((df['Vegetables Expenditure'] + df['Fruit Expenditure']) / df['Total Food Expenditure'])*100,0)
df['Other_Food_Share'] = np.round(((df['Bread and Cereals Expenditure'] + df['Vegetables Expenditure']) / df['Total Food Expenditure'])*100,0)
df['PerCapita_Protein'] = np.round(((df['Meat Expenditure'] + df['Total Fish and  marine products Expenditure']) / df['Total Number of Family members'])*100,0)

# Risk flag
df['Nutrition_Risk'] = ((df['Protein_Share'] < 0.2) & (df['Veggie_Fruit_Share'] < 0.1)).astype(int)

# Groupings
df['Income_Quintile'] = pd.qcut(df['Total Household Income'], 5, labels=['Q1 (Poorest)', 'Q2', 'Q3', 'Q4', 'Q5 (Richest)'])
df['Has_Under5'] = (df['Members with age less than 5 year old'] > 0).astype(int)
df['Agri_Household'] = df['Agricultural Household indicator'].astype(int)
df
```

```python
# Compute average rice share by income group
avg_rice = df.groupby('Income_Quintile')['Rice_Share'].mean()

# Plot
plt.figure(figsize=(8, 5))
avg_rice.plot(kind='bar', color='saddlebrown')
plt.title('Average % of Food Budget Spent on Rice by Income Group')
plt.ylabel('Rice Share')
plt.xlabel('Income Quintile')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Print insight
print(f"Poorest group (Q1) spends {avg_rice.iloc[0]:.1%} on rice.")
print(f"Richest group (Q5) spends {avg_rice.iloc[-1]:.1%} on rice.")
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='Has_Under5', y='PerCapita_Protein', palette='Set2')
plt.xticks([0, 1], ['No Child <5', 'Has Child <5'])
plt.title('Households with Young Children Get LESS Protein Per Person')
plt.ylabel('Per Capita Protein Spending (PHP)')
plt.show()
```

```python
region_stats = df.groupby('Region')[['Rice_Share', 'Protein_Share']].mean().reset_index()
fig, ax = plt.subplots(1, 2, figsize=(16, 6))
sns.barplot(data=region_stats, x='Region', y='Rice_Share', ax=ax[0], color='tan')
sns.barplot(data=region_stats, x='Region', y='Protein_Share', ax=ax[1], color='olive')
for a in ax: a.tick_params(axis='x', rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Income_Quintile', y='Veggie_Fruit_Share', palette='Blues')
plt.title('Even Middle-Income Households Reduce Fruit & Vegetable Spending')
plt.ylabel('Share of Food Budget on Fruits & Vegetables')
plt.show()
```

<h2>✅ ACT 1: DESCRIBING THE CRISIS</h2>
<p><em>What does “affording food” really look like?</em></p>

<div class="connection">
    🔗 Connection to ₱500 Noche Buena: If families can’t afford balanced meals daily, how can they afford a festive one?
</div>

<h3>D1. The Rice Trap: Staple Dominance in Tight Budgets</h3>
<ul>
    <li><strong>Metric:</strong> Rice_Share = Total Rice Expenditure / Total Food Expenditure</li>
    <li><strong>Insight:</strong> Low-income households spend over 60% of their food budget on rice—leaving almost nothing for protein or produce. Above ₱30,000/month income, rice share drops sharply.</li>
    <li><strong>Visualization:</strong> Scatter plot of Total Household Income (log scale) vs. Rice_Share and Protein_Share</li>
    <li><strong>Analytics Type:</strong> Descriptive (bivariate distribution, trend observation)</li>
</ul>
<p class="insight-quote">“When officials imagine a ₱500 Noche Buena with ‘rice and tinapa,’ they forget that for many, rice is already 70% of every meal—even without tinapa.”</p>

<h3>D2. Children Don’t Get Priority—They Get Less Per Person</h3>
<ul>
    <li><strong>Metric:</strong> PerCapita_Protein = (Meat Expenditure + Fish Expenditure) / Total Number of Family members</li>
    <li><strong>Insight:</strong> Households with children under 5 spend less on protein per person—nutrition is diluted, not prioritized.</li>
    <li><strong>Visualization:</strong> Boxplot comparing PerCapita_Protein by presence of young children</li>
    <li><strong>Analytics Type:</strong> Descriptive (group comparison, central tendency and spread)</li>
</ul>
<p class="insight-quote">“A Noche Buena for a family with three kids under five isn’t a luxury—it’s a financial emergency.”</p>

<h3>D3. Regional Nutrition Divide</h3>
<ul>
    <li><strong>Metric:</strong> Mean Rice_Share and Protein_Share by Region</li>
    <li><strong>Insight:</strong> ARMM, Eastern Visayas, and Bicol show extreme rice dependence (>65%) and critically low protein (<15%)—proving food inequality is geographic.</li>
    <li><strong>Visualization:</strong> Side-by-side regional bar charts</li>
    <li><strong>Analytics Type:</strong> Descriptive (geospatial aggregation, cross-regional comparison)</li>
</ul>
<p class="insight-quote">“In regions where ₱500 buys two days of basic meals, Noche Buena is a dream—not a budget line.”</p>

<h3>D4. Middle-Income Families Cut Back Too</h3>
<ul>
    <li><strong>Metric:</strong> Veggie_Fruit_Share = (Vegetables Expenditure + Fruit Expenditure) / Total Food Expenditure</li>
    <li><strong>Insight:</strong> Even Q3–Q4 income households reduce fruit/vegetable spending—nutritional stress spans beyond the poor.</li>
    <li><strong>Visualization:</strong> Boxplot across income quintiles</li>
    <li><strong>Analytics Type:</strong> Descriptive (stratified distribution by socioeconomic tier)</li>
</ul>
<p class="insight-quote">“The ₱500 myth assumes families have ‘extra’ to spend on holidays. But most are already cutting corners daily.”</p>

<h2>🔍 ACT 2: DIAGNOSING THE ROOTS</h2>
<p><em>Why can’t families afford a real Noche Buena?</em></p>

<div class="connection">
    🔗 Connection: The ₱500 claim ignores the structural trade-offs that define Filipino food budgets.
</div>

<h3>Dg1. Every Peso on Rice = Less for Meat</h3>
<ul>
    <li><strong>Metric:</strong> 
        <ul>
            <li>Pearson correlation between <code>Rice_Share</code> and <code>Protein_Share</code> (typically <i>r</i> ≈ –0.6)</li>
            <li>Regression coefficient of <code>Meat Expenditure</code> on <code>Total Rice Expenditure</code>, controlling for income</li>
        </ul>
    </li>
    <li><strong>Insight:</strong> 
        <ul>
            <li>Strong negative correlation confirms rice and protein compete for budget share.</li>
            <li>Regression shows: a ₱100 increase in rice spending reduces meat spending by ₱40–₱70—even after accounting for total income.</li>
        </ul>
    </li>
    <li><strong>Visualization:</strong> 
        <ul>
            <li>Correlation heatmap of food budget shares</li>
            <li>Scatter plot of actual vs. predicted meat expenditure</li>
        </ul>
    </li>
    <li><strong>Analytics Type:</strong> Diagnostic (correlation + multivariate regression)</li>
</ul>
<p class="insight-quote">“You can’t ‘just add tinapa’ when every extra peso on rice means less for protein.”</p>

<h3>Dg2. Bigger Families = Thinner Slices of the Pie</h3>
<ul>
    <li><strong>Metric:</strong> 
        <ul>
            <li>Pearson correlation between <code>Total Number of Family members</code> and <code>PerCapita_Protein</code></li>
            <li>Stratified correlation by income quintile (e.g., <i>r</i> = –0.62 in Q1 vs. –0.18 in Q5)</li>
        </ul>
    </li>
    <li><strong>Insight:</strong> 
        <ul>
            <li>Overall negative correlation, but much stronger among the poorest—proving crowding hits vulnerable households hardest.</li>
        </ul>
    </li>
    <li><strong>Visualization:</strong> 
        <ul>
            <li>Scatter plot of family size vs. per capita protein, colored by income quintile, with correlation coefficients annotated</li>
        </ul>
    </li>
    <li><strong>Analytics Type:</strong> Diagnostic (correlation with subgroup stratification)</li>
</ul>
<p class="insight-quote">“A ₱500 Noche Buena for a family of 8? That’s ₱62.50 per person—less than a single serving of fish.”</p>

<h3>Dg3. Agricultural Households Grow Food—But Not Diversity</h3>
<ul>
    <li><strong>Metric:</strong> 
        <ul>
            <li>Point-biserial correlation (via Pearson) between <code>Agri_Household</code> (binary) and <code>Veggie_Fruit_Share</code> (typically <i>r</i> ≈ –0.25)</li>
        </ul>
    </li>
    <li><strong>Insight:</strong>

            <li>Negative correlation confirms agricultural households consistently spend less on fruits and vegetables—likely due to limited dietary diversity in own production.</li>
        </ul>
    </li>
    <li><strong>Visualization:</strong> 
        <ul>
            <li>Boxplot of <code>Veggie_Fruit_Share</code> by agricultural status, with correlation value in caption</li>
        </ul>
    </li>
    <li><strong>Analytics Type:</strong> Diagnostic (correlation between binary and continuous variable)</li>
</ul>
<p class="insight-quote">“Farmers feed the nation—but their own tables lack variety. Noche Buena for them isn’t about money—it’s about access to diverse food.”</p>
